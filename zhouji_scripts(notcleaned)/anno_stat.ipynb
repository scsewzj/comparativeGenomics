```python
import os
from Bio import SeqIO
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------------------
# config
# ----------------------------

feature_files = {
    "CDS": "Oryza_sativa_cds.fa",
    "mRNA": "Oryza_sativa_mrna.fa",
    "protein": "Oryza_sativa_protein.fa"
}

"""
feature_files = {
    "CDS": "Oryza_brachyantha_cds.fa",
    "mRNA": "Oryza_brachyantha_mrna.fa",
    "protein": "Oryza_brachyantha_protein.fa"
}
"""

output_dir = "./plots/Oryza_sativa"
os.makedirs(output_dir, exist_ok=True)

# ----------------------------
# read lengths
# ----------------------------
length_data = []
for feature, file in feature_files.items():
    if os.path.exists(file):
        lengths = [len(rec.seq) for rec in SeqIO.parse(file, "fasta")]
        for l in lengths:
            length_data.append({"Feature": feature, "Length": l})
    else:
        print(f"Warning: {file} not found, skipped.")

df = pd.DataFrame(length_data)

# ----------------------------
# histogram
# ----------------------------
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x="Length", hue="Feature", bins=50, element="step", stat="count", multiple="stack")
plt.title("Feature Length Distribution")
plt.xlabel("Length (bp)")
plt.ylabel("Count")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.legend(feature_files.keys())
plt.savefig(os.path.join(output_dir, "feature_length_histogram.png"))
plt.close()

# ----------------------------
# boxplot
# ----------------------------
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="Feature", y="Length")
plt.xticks(rotation=45)
plt.title("Feature Length Boxplot")
plt.ylabel("Length (bp)")
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "feature_length_boxplot.png"))
plt.close()

print(f"Plots saved to {output_dir}")
```

```python
import pandas as pd
import numpy as np
from plotnine import *
from scipy.stats import mannwhitneyu

# -----------------------------
#  DataFrame
# df_cds, df_protein, df_mrna
#  DataFrame 'length'  'genome'
# -----------------------------

#  feature 
df_mrna = pd.DataFrame({
    "feature": ["mRNA"] * (len(os_lengths_mrna) + len(ob_lengths_mrna)),
    "length": os_lengths_mrna + ob_lengths_mrna,
    "genome": ["Oryza_sativa"] * len(os_lengths_mrna) + ["Oryza_brachyantha"] * len(ob_lengths_mrna)
})
df_protein = pd.DataFrame({
    "feature": ["protein"] * (len(os_lengths_protein) + len(ob_lengths_protein)),
    "length": os_lengths_protein + ob_lengths_protein,
    "genome": ["Oryza_sativa"] * len(os_lengths_protein) + ["Oryza_brachyantha"] * len(ob_lengths_protein)
})
df_cds = pd.DataFrame({
    "feature": ["CDS"] * (len(os_lengths_cds) + len(ob_lengths_cds)),
    "length": os_lengths_cds + ob_lengths_cds,
    "genome": ["Oryza_sativa"] * len(os_lengths_cds) + ["Oryza_brachyantha"] * len(ob_lengths_cds)
})

df_exons = pd.DataFrame({
    "feature": ["exon"] * (len(os_lengths_exons) + len(ob_lengths_exons)),
    "length": os_lengths_exons + ob_lengths_exons,
    "genome": ["Oryza_sativa"] * len(os_lengths_exons) + ["Oryza_brachyantha"] * len(ob_lengths_exons)
})

#  DataFrame
df = pd.concat([df_cds, df_protein, df_mrna, df_exons], ignore_index=True)

# -----------------------------
#  feature  Mann-Whitney U test
# -----------------------------
features = ['CDS', 'protein', 'mRNA', 'exon']
annot_list = []

palette_0 = {
    "Oryza_sativa": "#A3D5D3",        # 
    "Oryza_brachyantha": "#F7B7A3",   # 
    #"Oryza_example3": "#FFD6A5"       # 
}

palette_1 = {
    "Oryza_sativa": "#B790DA",       # 
    "Oryza_brachyantha": "#5E9ACE",  # 
    #"Oryza_example3": "#F67280"      # 
}

palette_2 = {
    #"Oryza_sativa": "#82C0CC",       # 
    "Oryza_brachyantha": "#B6D7A8",  # 
    "Oryza_sativa": "#F3E0BE"      # 
}

#palette_all = [palette_0, palette_1, palette_2]

for feat in features:
    os_vals = df[(df['feature']==feat) & (df['genome']=='Oryza_sativa')]['length']
    ob_vals = df[(df['feature']==feat) & (df['genome']=='Oryza_brachyantha')]['length']
    
    stat, p = mannwhitneyu(os_vals, ob_vals, alternative='two-sided')
    
    if p < 0.001:
        sig = '***'
    elif p < 0.01:
        sig = '**'
    elif p < 0.05:
        sig = '*'
    else:
        sig = ''
    
    y_pos = df[df['feature']==feat]['length'].max() * 1.1
    annot_list.append({"feature": feat, "y": y_pos, "label": sig})

annot_df = pd.DataFrame(annot_list)

# -----------------------------
# 
# -----------------------------


# -----------------------------
# 
# -----------------------------

(
    ggplot(df, aes(x='feature', y='length', fill='genome'))
    + geom_violin(trim=False)
    + stat_summary(fun_y=np.median, geom="point", color="black", size=2,
                   position=position_dodge(width=0.9))
    + stat_summary(fun_ymin=lambda x: np.quantile(x,0.25),
                   fun_ymax=lambda x: np.quantile(x,0.75),
                   geom="errorbar", width=0.2, color="black",
                   position=position_dodge(width=0.9))
    + geom_text(
        data=annot_df,
        mapping=aes(x='feature', y='y', label='label'),
        inherit_aes=False,
        ha='center',
        size=10
    )
    + scale_y_log10()
    + scale_fill_manual(values=palette_2)
    + theme_bw()
    + theme(
        figure_size=(8,5),
        text=element_text(size=12)
    )
)
```

```python
import gffutils

# 1. 
db_ob = gffutils.create_db("Oryza_brachyantha.v1.4b.41.gff3", dbfn="ob.db", force=True, keep_order=True, merge_strategy='merge', sort_attribute_values=True)

exons_complete_ob = []
intron_complete_ob = []
# 2.  transcript
for tx in db_ob.features_of_type("transcript"):
    exons = list(db_ob.children(tx, featuretype="exon", order_by="start"))
    exons_complete_ob.extend(exons)
    # exon  intron
    for i in range(len(exons)-1):
        intron_start = exons[i].end + 1
        intron_end = exons[i+1].start - 1
        if intron_end >= intron_start:
            print(f"Intron: {tx.id}, {exons[i].chrom}, {intron_start}-{intron_end}")
            intron_complete_ob.append((tx.id, exons[i].chrom, intron_start, intron_end))
```

```python
from pyfaidx import Fasta
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# -----------------------------
# 1. 
# -----------------------------
genome = Fasta("./Oryza_sativa_japonica.IRGSP-1.0.dna.toplevel.fa")

window_size = 1_000_000    # 1 Mb
step_size   = 200_000      # 0.2 Mb 

results = []

# -----------------------------
# 2.  GC
# -----------------------------
for chrom in genome.keys():
    seq_len = len(genome[chrom])
    for start in range(0, seq_len, step_size):
        end = min(start + window_size, seq_len)
        seq = genome[chrom][start:end].seq.upper()
        gc_count = seq.count("G") + seq.count("C")
        gc_frac = gc_count / len(seq)
        results.append({
            "chrom": chrom,
            "start": start,
            "end": end,
            "gc_frac": gc_frac
        })

gc_df = pd.DataFrame(results)

# -----------------------------
# 3. 
# -----------------------------
chroms = sorted(gc_df['chrom'].unique())
chrom_offsets = {}
total_len = 0

#  x 
for chrom in chroms:
    chrom_len = gc_df[gc_df['chrom']==chrom]['end'].max()
    chrom_offsets[chrom] = total_len
    total_len += chrom_len

# 
gc_df['abs_pos'] = gc_df.apply(lambda row: row['start'] + chrom_offsets[row['chrom']], axis=1)

# 
plt.figure(figsize=(20,5))
palette = sns.color_palette("tab20", len(chroms))
for i, chrom in enumerate(chroms):
    chrom_data = gc_df[gc_df['chrom']==chrom]
    plt.plot(chrom_data['abs_pos']/1e6, chrom_data['gc_frac'], label=chrom, color=palette[i])

plt.xlabel("Genome position (Mb)")
plt.ylabel("GC fraction")
plt.title("Genome-wide GC content (1Mb sliding window)")
#plt.legend(ncol=2, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
