```python
import pandas as pd
import numpy as np
```

```python
specis_prefix = 'os'
with open('circos_japonica_te_density.txt', 'r') as file:
    lines = file.readlines()
data = []
for line in lines:
    data.append(line.strip().split('\t'))
import pandas as pd
df = pd.DataFrame(data, columns=['chrom', 'start', 'end', 'density'])
df['start'] = df['start'].astype(int)
df['end'] = df['end'].astype(float)
df['end']=df['end'].astype(int)
df['density'] = df['density'].astype(float)
for i in range(len(df)):
    try :
        df['chrom'][i] = int(df['chrom'][i])
    except:
        df['chrom'][i] = np.nan
df.dropna(subset=['chrom'], inplace=True)
for i in range(len(df)):
    df.iloc[i, 0] = specis_prefix + str(int(df.iloc[i, 0]))
df.to_csv('os_te_density_processed', sep='\t', index=False, header=False)
```

```python
from collections import defaultdict

gff_file = "../anno_stat/Oryza_sativa.IRGSP-1.0.41.gff3"
karyotype_file = "os_karyotype.txt"

chrom_lengths = defaultdict(int)

# 读取每条染色体的最大坐标
with open(gff_file) as f:
    for line in f:
        if line.startswith("#"):
            continue
        parts = line.strip().split("\t")
        chrom = parts[0]
        start = int(parts[3])
        end   = int(parts[4])
        if end > chrom_lengths[chrom]:
            chrom_lengths[chrom] = end

# 写入 karyotype 文件
with open(karyotype_file, "w") as out:
    for i, (chrom, length) in enumerate(chrom_lengths.items(), 1):
        # chr - <ID> <label> <start> <end> <color>
        out.write(f"chr - {chrom} {chrom} 0 {length} grey\n")
```

```python
# homology
with open('homology.txt') as file:
    lines = file.readlines()
for i in range(len(lines)):
    lines[i] = lines[i].strip().split(' ')
data = [(lines[i],lines[i+1]) for i in range(0, len(lines), 2)]

ob_intra=[]
os_intra=[]
inter = []
for pair in data:
    if pair[0][1].startswith('ob') and pair[1][1].startswith('ob'):
        ob_intra.append(pair)
    elif pair[0][1].startswith('os') and pair[1][1].startswith('os'):
        os_intra.append(pair)
    else:
        inter.append(pair)
with open('ob_intra_homo.txt', 'w') as out:
    for i, j in ob_intra:
        out.write(f"{i[0]} {i[1]} {i[2]} {i[3]} {i[4]}\n")
        out.write(f"{j[0]} {j[1]} {j[2]} {j[3]} {j[4]}\n")

with open('os_intra_homo.txt', 'w') as out:
    for i, j in os_intra:
        out.write(f"{i[0]} {i[1]} {i[2]} {i[3]} {i[4]}\n")
        out.write(f"{j[0]} {j[1]} {j[2]} {j[3]} {j[4]}\n")
        
with open('inter_homo.txt', 'w') as out:
    for i, j in inter:
        out.write(f"{i[0]} {i[1]} {i[2]} {i[3]} {i[4]}\n")
        out.write(f"{j[0]} {j[1]} {j[2]} {j[3]} {j[4]}\n")
```
